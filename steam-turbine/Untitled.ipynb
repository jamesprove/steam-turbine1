```python
data_train = pd.read_table("/Users/nizhenming/Desktop/machine learning/steam turbine/zhengqi_train.txt",encoding='utf-8')
data_test = pd.read_table("/Users/nizhenming/Desktop/machine learning/steam turbine/zhengqi_test.txt",encoding='utf-8')
```

```python
t = data_train.corr()
t
```

```python
drop_columns = t[(np.abs(t["target"])<0.15)].index
data_train.drop(columns=drop_columns,inplace=True)
data_test.drop(columns=drop_columns,inplace=True)
```

```python
t = data_train.corr()
t
```

```python
from sklearn.model_selection import train_test_split

X = data_train.iloc[:,:-1]
y = data_train['target']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression   
from sklearn.metrics import mean_squared_error
clfL = LinearRegression()

clfL.fit(X_train,y_train)

y_true, y_pred = y_test, clfL.predict(X_test)

print(mean_squared_error(y_true, y_pred))

ans_Liner = clfL.predict(data_test)
print(ans_Liner.shape)
```

```python
result = pd.DataFrame(ans_Liner)
result.to_csv(r'./Liners.txt', header=None, index=None, mode='a', sep='\t')

```
